# 4) Cleaning the dataset

```python
raw_df.drop(['Zomato URL'],axis=1,inplace=True)
```

```python
raw_df.head()
```

```python
raw_df['Location'].value_counts()
```

```python
raw_df['Dining Rating'].value_counts()
```

```python
raw_df['Dining Rating'].replace(to_replace=["None"],value="0",inplace=True)
```

```python
raw_df['Dining Rating Count'].value_counts()
```

```python
raw_df['Dining Rating Count'].replace(to_replace=['Does not offer Dining','Not enough Dining Reviews','None'],
                                      value='0',inplace=True)
```

```python
raw_df['Delivery Rating'].value_counts()
```

```python
raw_df['Delivery Rating'].replace(to_replace=["None"],value="0",inplace=True)
```

```python
raw_df['Delivery Rating Count'].value_counts()
```

```python
raw_df['Delivery Rating Count'].replace(to_replace=['Not enough Delivery Reviews','Does not offer Delivery','None'],
                                      value='0',inplace=True)
```

```python
raw_df['Dining Rating']=raw_df['Dining Rating'].astype('float64')
raw_df["Dining Rating Count"]=raw_df['Dining Rating Count'].astype('int64')
raw_df['Delivery Rating']=raw_df['Delivery Rating'].astype('float64')
raw_df['Delivery Rating Count']=raw_df['Delivery Rating Count'].astype('int64')
```

```python
raw_df.info()
```

# removing the restaurant records in rating is 0

```python
useless_data=(raw_df['Dining Rating'])==0|(raw_df['Delivery Rating']== 0)
raw_df[useless_data]
```

```python
raw_df=raw_df[~useless_data]
```

```python
useless_data=(raw_df['Dining Rating Count'])==0|(raw_df['Delivery Rating Count']== 0)
raw_df[useless_data]
```

```python
raw_df=raw_df[~useless_data]
```

```python
raw_df.shape
```

```python
raw_df.head(10)
```

# removing duplicate records

```python
raw_df[raw_df.duplicated()]
```

```python
raw_df=raw_df.drop_duplicates()
```

```python
affprice=zomato_df['Price for 2'].max()
affprice/4
```

```python
affresta=zomato_df[['Name of Restaurant','Location',"Cuisine",'Price for 2']]
affresta=affresta[affresta['Price for 2']<1250]
affresta.sort_values(by='Price for 2',inplace=True)
affresta.head(10)
```

```python
highestrating=zomato_df[['Name of Restaurant','Location',"Cuisine",'Price for 2','Delivery Rating']]
highestrating=highestrating[highestrating['Delivery Rating']>4.5]
highestrating.sort_values(by='Delivery Rating',inplace=True)
highestrating
```

```python
highest_aff=pd.merge(highestrating,affresta,how='inner',on=['Name of Restaurant','Location','Price for 2','Cuisine'])
highest_aff=highest_aff[['Name of Restaurant','Price for 2','Location','Cuisine','Delivery Rating']]
highest_aff.sort_values(by='Price for 2',inplace=True)
highest_aff
```
